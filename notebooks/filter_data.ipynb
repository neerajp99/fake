```python
# Import libraries 
import pandas as pd 
import numpy as np 
```

```python
# Get the train datasets 
train_headings = pd.read_csv('../data/fnc-1/train_stances.csv')
train_bodies = pd.read_csv('../data/fnc-1/train_bodies.csv')
# Get the test datasets  
test_headings = pd.read_csv('../data/fnc-1/competition_test_stances.csv')
test_bodies = pd.read_csv('../data/fnc-1/competition_test_bodies.csv')
```

```python
# Temporary lists to store train values 
train_bodyID = list()
train_articleHeading = list()
train_articleBody = list()
train_articleStance = list()
```

```python
# Temporary lists to store test values 
test_bodyID = list()
test_articleHeading = list()
test_articleBody = list()
test_articleStance = list()
```

```python
# Form the training dataframe 
for body_value in train_bodies.index:
    body_id = train_bodies['Body ID'][body_value]
    current_heading = train_headings.loc[train_headings['Body ID'] == body_id]
    # Get the article body
    current_body_value = train_bodies['articleBody'][body_value]
    # Loop over the articles with the same body and fetch details and append to the lists
    for head_value in current_heading.index:
        current_heading_value = current_heading['Headline'][head_value]
        current_stance_value = current_heading['Stance'][head_value]
        # Append the values to the dataframe 
        train_bodyID.append(body_id)
        train_articleHeading.append(current_heading_value)
        train_articleBody.append(current_body_value)
        train_articleStance.append(current_stance_value)
```

```python
# Create an empty dataframe as the final train dataframe 
train_data = pd.DataFrame()
train_data['bodyId'] = train_bodyID
train_data['articleHeading'] = train_articleHeading
train_data['articleBody'] = train_articleBody
train_data['articleStance'] = train_articleStance
```

```python
# Form the testing dataframe 
for body_value in test_bodies.index:
    body_id = test_bodies['Body ID'][body_value]
    current_heading = test_headings.loc[test_headings['Body ID'] == body_id]
    # Get the article body
    current_body_value = test_bodies['articleBody'][body_value]
    # Loop over the articles with the same body and fetch details and append to the lists
    for head_value in current_heading.index:
        current_heading_value = current_heading['Headline'][head_value]
        current_stance_value = current_heading['Stance'][head_value]
        # Append the values to the dataframe 
        test_bodyID.append(body_id)
        test_articleHeading.append(current_heading_value)
        test_articleBody.append(current_body_value)
        test_articleStance.append(current_stance_value)
```

```python
# Create an empty dataframe as the final train dataframe 
test_data = pd.DataFrame()
test_data['bodyId'] = test_bodyID
test_data['articleHeading'] = test_articleHeading
test_data['articleBody'] = test_articleBody
test_data['articleStance'] = test_articleStance
```
